```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image, display

from IPython.display import Image, display




df=pd.read_csv('/kaggle/input/datasets/litonislam/hr-analytics-cleaned-and-processed-employee-dataset/HR_Data_Cleaned.csv')


ig, ax = plt.subplots(2, 2, figsize=(16, 12))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# পাই চার্ট: জেন্ডার ডিস্ট্রিবিউশন
gender_counts = df['Gender'].value_counts()
ax[0, 0].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', 
             startangle=140, colors=['#3498db', '#e84393'], 
             textprops={'fontsize': 12, 'fontweight': 'bold'})
ax[0, 0].set_title('Python: Gender Distribution', fontsize=14, fontweight='bold', pad=15)

# ডোনাট চার্ট: বিজনেস ট্রাভেল
travel_counts = df['BusinessTravel'].value_counts()
ax[0, 1].pie(travel_counts, labels=travel_counts.index, autopct='%1.1f%%', 
             startangle=140, colors=sns.color_palette('viridis', len(travel_counts)),
             wedgeprops={'width': 0.5}, textprops={'fontsize': 12, 'fontweight': 'bold'})
ax[0, 1].set_title('Python: Business Travel Frequency', fontsize=14, fontweight='bold', pad=15)

# বার চার্ট: ডিপার্টমেন্ট অনুযায়ী এমপ্লয়ি সংখ্যা
plt.subplot(2, 1, 2) 
ax_bar = sns.countplot(data=df, x='Department', hue='Department', palette='viridis', legend=False)
plt.title('Python: Employee Count by Department', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Department Name', fontsize=12, fontweight='bold')
plt.ylabel('Count', fontsize=12, fontweight='bold')

# বারের উপরে সংখ্যা (Data Labels) বসানো
for p in ax_bar.patches:
    ax_bar.annotate(f'{int(p.get_height())}', 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha='center', va='baseline', 
                   fontsize=12, color='black', xytext=(0, 5),
                   textcoords='offset points')
    
plt.savefig('hr_analytics_dashboard.png')
plt.tight_layout()
plt.show()






```
